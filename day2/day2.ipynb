 线性回归

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

X,y=make_regression(n_samples=1000,n_features=1,noise=20,random_state=114514)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,shuffle=False)
linear=LinearRegression().fit(X_train,y_train)

plt.scatter(X,y,c='b',s=60)
plt.plot(X,linear.predict(X),c='r')
plt.title('linear regression')

print(f"训练数据集得分：{linear.score(X_train,y_train)}")
print(f"测试数据集得分：{linear.score(X_test,y_test)}")
```

波士顿房价预测（二元线性回归）

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

filename='boston_house.csv'
table=pd.read_csv(filename)
#print(table)

features=table.columns.drop('MEDV')#获取列变量用于遍历
y=table['MEDV']
for feature in features:
        X=table[[feature]] #一个方括号返回的是一个series对象，两个括号返回的是一个dataframe对象

        X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,shuffle=True)
        linear=LinearRegression().fit(X_train,y_train)

        plt.figure()
        plt.scatter(X,y,c='b',s=60)
        plt.plot(X,linear.predict(X),c='r')
        plt.title(f'linear regression about MEDV and {feature}')
        plt.ylabel('MEDV')
        plt.xlabel(feature)
        plt.show(block=False)#让图显示同时不阻止程序运行
        print(f"对于MEDV和{feature}的线性拟合：")
        print(f"训练数据集得分：{linear.score(X_train,y_train)}")
        print(f"测试数据集得分：{linear.score(X_test,y_test)}\n")
```
